```python
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
```

```python
def DP(x):
    return np.cos(x*3)*np.sin(x) / (np.sqrt(2*np.pi*0.2**2)) * np.exp (-0.5*((x - 5 / 1.5)**2))/np.sqrt(1.150)
    

x_array = np.arange(-8,8,0.1)

area,err = integrate.quad(lambda x :DP(x)**2,-np.inf,np.inf)
media,err = integrate.quad(lambda x :x*DP(x)**2,-np.inf,np.inf)
media2,err = integrate.quad(lambda x :x**2*DP(x)**2,-np.inf,np.inf)
desvio = np.sqrt(media2-media**2)

print (f'área = {area :.3f}')
print (f'média = {media :.3f}')
print (f'desvio padrão = {desvio :.3f}')

plt.plot(x_array,DP(x_array)**2)
plt.title('Densidade de Probabilidade')
plt.xlabel('x')
plt.ylabel('$|\Psi(x)|^2$')
```

```python
#EXEMPLO 1
def DP_complex(x):
    r = (x-2)* np.exp (-0.5*((x + 2)**2))/np.sqrt(29.24549)
    return complex(r*np.cos(np.pi*(3/4)*x),r*np.sin(np.pi*(3/4)*x))
    

def PD(x):
    z = DP_complex(x)
    return z.real**2 + z.imag**2

pp=[]
for i in x_array:
    pp.append(PD(i))

area,err = integrate.quad(lambda x :PD(x),-np.inf,np.inf)
media,err = integrate.quad(lambda x :x*PD(x),-np.inf,np.inf)
media2,err = integrate.quad(lambda x :x**2*PD(x),-np.inf,np.inf)
delta_x = np.sqrt(media2-media**2)

print (f'área = {area :.5f}')
print (f'média = {media :.3f}')
print (f'desvio padrão = {delta_x :.3f}')

plt.plot(x_array, pp)
plt.title('Densidade de Probabilidade')
plt.xlabel('x')
plt.ylabel('$|\Psi(x)|^2$')

```

```python
def A(k):
    def integrando1(x):
        z = DP_complex(x)
        return z.real*np.cos(k*x)-z.imag*np.sin(-k*x)
    
    def integrando2(x):
        z = DP_complex(x)
        return z.real*np.sin(-k*x)+z.imag*np.cos(k*x)   
    
    A1,err1 = integrate.quad(integrando1,-np.inf,np.inf)
    A2,err2 = integrate.quad(integrando2,-np.inf,np.inf) 
    return complex(A1/(np.sqrt(2*np.pi)),A2/(np.sqrt(2*np.pi)))

def PF(k):
    z = A(k)
    return (z.real**2 + z.imag**2)

dk = 0.1
FF=[]
for i in x_array:
    FF.append(PF(i))
    
area = np.sum(FF)*dk
media = np.sum(x_array*FF)*dk
media2 = np.sum(x_array**2*FF)*dk
delta_k = np.sqrt(media2-media**2)

print (f'área = {area :.3f}')
print (f'média = {media :.3f}')
print (f'desvio padrão = {delta_k :.3f}')

    
plt.plot(x_array,FF)
plt.title('Densidade de Probabilidade k')
plt.xlabel("k")
plt.ylabel('$|A(k)|^2$')
```

```python
plt.figure(figsize=(10,12))

plt.subplot(221)
plt.plot(x_array,pp)
plt.xlabel('x') 
plt.ylabel('$|\Psi(x)|^2$')
plt.legend()

plt.subplot(222)
plt.plot(x_array,FF)
plt.xlabel('k') 
plt.ylabel('$|A(k)|^2$')

plt.show()
p = delta_x*delta_k
print(f'dx dk = {p :.3f}')
```
